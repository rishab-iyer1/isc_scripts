```python
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate variance
def calculate_variance(ratings):
    n = len(ratings)
    mean = np.mean(ratings)
    variance = np.sum((ratings - mean) ** 2) / n
    return variance

# Function to calculate absolute difference from 0.5
def calculate_absolute_difference(ratings):
    mean = np.mean(ratings)
    abs_diff = 0.5 - abs(mean - 0.5)
    return abs_diff

def generate_ratings(length):
    """
    Generate an ndarray of ratings, ranging from all 0s to all 1s.
    
    Parameters:
    - length (int): The number of elements in each rating array.
    
    Returns:
    - ndarray: A 2D array where each row is a rating combination.
    """
    num_rows = length + 1
    ratings = np.array([np.concatenate((np.ones(i), np.zeros(length - i))) for i in range(num_rows)])
    return ratings

length = 100
ratings = generate_ratings(length)
print(ratings, ratings.shape)

# Calculate variance and absolute difference
variance = [calculate_variance(r) for r in ratings]
abs_diff = [calculate_absolute_difference(r) for r in ratings]

# Print results
# print(f"Ratings: {ratings}")
# print(f"Variance: {variance:.4f}")
# print(f"Absolute Difference from 0.5: {abs_diff:.4f}")

mean = np.mean(ratings, axis=1)
plt.plot(mean, variance, label='variance')
plt.plot(mean, abs_diff, label='abs_diff')
plt.legend()
plt.show()
```
